```python
#Import relevant packages for analyses 
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
import xarray as xr   
import momlevel as ml #import momlevel for SST calc
import cftime
import cmocean 
from cartopy import crs as ccrs, feature as cfeature 

#Define figures to be of certain size
%matplotlib inline                                     
%config InlineBackend.figure_format='retina' 
plt.rcParams['figure.figsize'] = 12,6       
```

```python
### Import CESMB Ocean model Data

### Pre EOT Ocean data
cesmb_sst_PREEOT= xr.open_dataset('38Ma_4PIC_Ocean_temp.nc')
cesmb_sst_um_bil_PREEOT=xr.open_dataset('38Ma_4PIC_Ocean_temp_um_bil.nc')

### Post EOT Ocean data
cesmb_sst_POSTEOT= xr.open_dataset('38Ma_2PIC_Ocean_temp.nc')
cesmb_sst_um_bil_POSTEOT=xr.open_dataset('38Ma_2PIC_Ocean_temp_um_bil.nc')
```

## CESM-B Raw Data Analysis

### SST Data

```python
#Plot CESM-B PreEot SST data
cesmb_sst_PREEOT.TEMP.plot(cmap=cmocean.cm.balance)
plt.title("CESM-B: PRE-EOT Global SST", fontweight="bold",  fontsize=15) #Set title 
```

```python
#Plot CESM-B PostEot SST data
cesmb_sst_POSTEOT.TEMP.plot(cmap=cmocean.cm.balance)
plt.title("CESM-B: POST-EOT Global SST", fontweight="bold",  fontsize=15) #Set title 
```

```python
#Find SST difference of PostEOT and PreEot SST in CESM-B data
cesmb_sst_diff_POST_PRE=cesmb_sst_POSTEOT - cesmb_sst_PREEOT
```

```python
cesmb_sst_diff_POST_PRE
```

```python
#Plot SST difference of PostEOT and PreEot
cesmb_sst_diff_POST_PRE.TEMP.plot(cmap=cmocean.cm.balance)
plt.title("CESM-B: PRE-EOT - POST-EOT Global SST", fontweight="bold",  fontsize=15) #Set title 
```

### SAT Data

```python
### Import CESMB Atmosphere Model Data

### Pre EOT Atmos data
cesmb_atm_PREEOT= xr.open_dataset('38Ma_4PIC_Atm_temp.nc')
cesmb_atm_um_bil_PREEOT=xr.open_dataset('38Ma_4PIC_Atm_temp_um_bil.nc')

### Post EOT Atmos data
cesmb_atm_POSTEOT= xr.open_dataset('38Ma_2PIC_Atm_temp.nc')
cesmb_atm_um_bil_POSTEOT=xr.open_dataset('38Ma_2PIC_Atm_temp_um_bil.nc')
```

```python
#Plot PostEot SST data
cesmb_atm_POSTEOT.TS.plot(cmap=cmocean.cm.balance)
plt.title("CESM-B: POST-EOT Global SAT", fontweight="bold",  fontsize=15) #Set title 
```

```python
#Find SAT difference of PostEOT and PreEOT
cesmb_atm_diff_POST_PRE=cesmb_atm_POSTEOT - cesmb_atm_PREEOT
```

```python
#Find SST difference between PostEOT and PreEOT
cesmb_diff_POST_PRE_um_bil=cesmb_sst_um_bil_POSTEOT - cesmb_sst_um_bil_PREEOT
```

```python
#Plot SST difference between PostEOT and PreEOT on common grid
cesmb_diff_POST_PRE_um_bil.temp.plot(cmap=cmocean.cm.balance)
plt.title("CESM-B: POST-EOT - PRE-EOT Global SST (Common Grid)", fontweight="bold",  fontsize=15) #Set title 
```

## Projections

```python
#Define projection type 
subplot_kws=dict(projection=ccrs.Robinson(),
                 facecolor='black') #Set nackground to black

#Setup 2 figures of SAT and SST on Robinson projections 
fig, (ax1, ax2 ) = plt.subplots(ncols=1,
                               nrows=2,
                               figsize=(8,10),
                               subplot_kw=subplot_kws)

cesmb_atm_diff_POST_PRE.TS.plot.contourf(ax=ax1,  #Plot CESM-B SAT change across the EOT boundary, smooth out contours 
                              cmap=cmocean.cm.balance,
                              extend='both',vmin=-8,vmax=8,
                            transform=ccrs.PlateCarree(),levels=60)

cesmb_atm_diff_POST_PRE.TS.plot.contour(ax=ax1,  #Add boundaries between temp changes
                              colors='white',linestyles='-',linewidths=0.1,
                              extend='both',vmin=-8,vmax=8,
                            transform=ccrs.PlateCarree(),levels=60,)
ax1.set_title("CESM-B: Global SAT Change Across EOT", fontweight="bold",  fontsize=10) #Set title

ax1.coastlines() #Plot modern day coastlines


cesmb_diff_POST_PRE_um_bil.temp.plot.contourf(ax=ax2, #Plot CESM-B SST change across the EOT boundary, smooth out contours 
                              cmap=cmocean.cm.balance,
                              extend='both',vmin=-5,vmax=5,
                            transform=ccrs.PlateCarree(),levels=60,)

cesmb_diff_POST_PRE_um_bil.temp.plot.contour(ax=ax2,  #Add boundaries between temp changes
                              colors='white',linestyles='-',linewidths=0.1,
                              extend='both',vmin=-5,vmax=5,
                            transform=ccrs.PlateCarree(),levels=60,)
ax2.set_title("CESM-B: Global SST Change Across EOT", fontweight="bold",  fontsize=10) #Set title
```

#### **Figure 1.** Global SAT change, top, and global SST change, bottom, before and after the EOT using data generated from the CESM-B model. Overall, there is a decreasing trend in both SAT and SST. SAT saw a decrease in temperature ranging from 1.5 to 7 degrees Celsius, averaging 3 degrees. The poles and atmosphere above landmasses saw the greatest decrease in temperature. SST saw a decrease in temperature ranging from 1 to 5 degrees, averaging around 3 degrees. The northern hemisphere saw a greater cooling than the southern hemisphere. Model data was run and compiled by Baatsen et al. (2020).
