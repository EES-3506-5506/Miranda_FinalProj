## Projections

```python
#Define projection type
subplot_kws=dict(projection=ccrs.Robinson(),
                 facecolor='black') #Set blckground to black

#Setup 2 figures of SAT and SST on Robinson projections 
fig, (ax1, ax2 ) = plt.subplots(ncols=1,
                               nrows=2,
                               figsize=(8,10),
                               subplot_kw=subplot_kws)

had_atm_cg_diff_POST_PRE.temp_mm_1_5m.plot.contourf(ax=ax1, #Plot HadCM3L SAT change across the EOT boundary, smooth out contours 
                              cmap=cmocean.cm.balance,
                              extend='both',vmin=-15,vmax=15,
                            transform=ccrs.PlateCarree(),levels=40,)


had_atm_cg_diff_POST_PRE.temp_mm_1_5m.plot.contour(ax=ax1,#cbar_kwargs={"orientation": "horizontal", "shrink": 0.3, "aspect": 20, "label":"[W/m2]", "pad": 0.02},
                              colors='white',linestyles='-',linewidths=0.1, #Add boundaries between temp changes
                              extend='both',vmin=-15,vmax=15,
                            transform=ccrs.PlateCarree(),levels=40,)

ax1.set_title("HadCM3L: Global SAT Change Across EOT", fontweight="bold",  fontsize=10) #Set title
ax1.coastlines() #Lot modern day coastlines


had_difference_sst_POST_PRE.temp_ym_dpth.plot.contourf(ax=ax2, #Plot HadCM3L SST change across the EOT boundary, smooth out contours
                              cmap=cmocean.cm.balance,
                              extend='both',vmin=-4,vmax=4,
                            transform=ccrs.PlateCarree(),levels=40,)

had_difference_sst_POST_PRE.temp_ym_dpth.plot.contour(ax=ax2,  #Add boundaries between temp changes
                              colors='white',linestyles='-',linewidths=0.1,
                              extend='both',vmin=-4,vmax=4,
                            transform=ccrs.PlateCarree(),levels=40,)
ax2.set_title("HadCM3L: Global SST Change Across EOT", fontweight="bold",  fontsize=10) #Set title
```

#### **Figure 1.** Global SAT change, top, and global SST change, bottom, before and after the EOT using data generated from the HadCM3L model. Overall, there is a decreasing trend in both SAT and SST. SAT saw a decrease in temperature ranging from 0.5 to 15 degrees Celsius, averaging 3.5 degrees. SST saw a decrease in temperature ranging from 0.5 to 4 degrees, averaging around 2.5 degrees. The northern hemisphere saw a greater cooling than the southern. SAT and SST had anomalous warming in western Antarctica. This is likely a limitation of the model.
